```
# DBSCAN
import pandas as pd
data = pd.read_csv('/content/Mall_Customers.csv')
```

```
data.head()
```

```
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=6, min_samples=7, metric='euclidean')
```

```
# X and Y
X = data.iloc[:,[3,4]].values
Y = data.iloc[:,-1].values
```

```
model = dbscan.fit(X)
label = model.labels_
print(label)
```
